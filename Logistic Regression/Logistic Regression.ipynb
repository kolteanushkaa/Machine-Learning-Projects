#### Business Usecases:- Build a model that predicts the speices of Iris flowers based on sepal length,sepal width,petal width and petal length

```python
###### Step-1 Data Gathering
path=r"https://raw.githubusercontent.com/sindhura-nk/Datasets/refs/heads/main/iris.csv"
import pandas as pd
df=pd.read_csv(path)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
X= df.drop(columns=["species"])
y=df["species"]
```

```python
X.head()
```

```python
y.head()
```

```python
#Feature Engineering: Data cleaning and data preprocessing

from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
pipeline= make_pipeline(SimpleImputer(strategy="mean"),
                        StandardScaler()
                        ).set_output(transform="pandas")
pipeline
```

```python
x_pipeline=pipeline.fit_transform(X)
display(x_pipeline.head())
display(X.head())
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_pipeline,y)
```

```python
model.score(x_pipeline,y)
```

```python
from sklearn.metrics import f1_score,classification_report,confusion_matrix
ypred=model.predict(x_pipeline)
f1 = f1_score(y, ypred, average="weighted")
class_report = classification_report(y, ypred)
conf_matrix = confusion_matrix(y, ypred)
```

```python
print(class_report)
```

```python
import joblib
joblib.dump(model,"model.joblib")
joblib.dump(pipeline,"pipeline.joblib")
```

```python
path2="https://raw.githubusercontent.com/sindhura-nk/Datasets/refs/heads/main/iris_sample.csv"
xnew=pd.read_csv(path2)
xnew.head()
```
